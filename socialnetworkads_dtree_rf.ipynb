```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
socialnetwork=pd.read_csv("/content/drive/MyDrive/SocialNetworkAds (1).csv")
socialnetwork
```

```python
socialnetwork['Gender']=[1 if each=='Male' else 2 for each in socialnetwork.Gender]
socialnetwork
```

```python
a=socialnetwork.drop(['User ID'],axis=1)
a
```

```python
y=a['Purchased']
x=a.drop(['Purchased'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=100,test_size=0.25)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(x_train,y_train)
```

```python
y_pred=dtree.predict(x_test)
y_pred
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100)
rf.fit(x_train,y_train)
```

```python
y_predi=rf.predict(x_test)
y_predi
```

```python
z_pred=dtree.predict(df)
z_pred
if z_pred==1:
  print("purchase")
else:
  print("not purchased")
```
