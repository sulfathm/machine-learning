```python
data1=data.drop(['Survived','Fare',"Ticket","Name","Sex","Embarked","Cabin"],axis=1)
data1.columns
```

```python
y=data1['Age']

x=data1.drop(['Age'],axis=1)
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
from sklearn import linear_model
lr=linear_model.LinearRegression()
```

```python
x_train
```

```python
x_test
```

```python
a=y_train.fillna(value=2)
a
```

```python
y_train
```

```python
b=y_test.fillna(value=2)
b


```

```python
model=lr.fit(x_train,a)
```

```python
predictions=model.predict(x_test)
print(predictions)
```

```python
from sklearn.metrics import mean_squared_error
print('MSE is :',mean_squared_error(b,predictions) )
```
