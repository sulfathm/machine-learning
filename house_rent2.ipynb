```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/drive/MyDrive/house_rent2.csv")
data
```

```python
a=data.drop(["Posted On","Floor","Area Type","Area Locality","City","Furnishing Status","Tenant Preferred","Point of Contact"],axis=1)
a
```

```python
y=a["Rent"]
x=a.drop(["Rent"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_predi=lr.predict(x_test)
y_predi
```
