```python
import pandas as pd
import numpy as np
```

```python
stroke=pd.read_csv("/content/drive/MyDrive/healthcare-dataset-stroke-data.csv")
stroke
```

```python
a=stroke.drop(['id','work_type','Residence_type'],axis=1)
```

```python
a
```

```python
b=a.fillna(value=22)
b
```

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
```

```python
b['gender']=label_encoder.fit_transform(b['gender'])
b['ever_married']=label_encoder.fit_transform(b['ever_married'])
b["smoking_status"]=label_encoder.fit_transform(b['smoking_status'])
```

```python
b
```

```python
y=b['stroke']
x=b.drop(['stroke'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=100,test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train,y_train)
```

```python
x_test
```

```python
y_predict=knn.predict(x_test)
y_predict
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(x_train,y_train)
```

```python
y_predict=dtree.predict(x_test)
y_predict
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100)
rf.fit(x_train,y_train)
```

```python
y_predict=rf.predict(x_test)
y_predict
```
