```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic=pd.read_csv("/content/drive/MyDrive/titanic.csv")
titanic
```

```python
a=titanic.drop(['Cabin','Embarked','Name',"Ticket","PassengerId","SibSp","Parch"],axis=1)
a
```

```python
a.isna().sum()
```

```python
a['Age']=a['Age'].fillna(a['Age'].median())
a
```

```python
a["Sex"]=[2 if each=='female' else 1 for each in a.Sex]
a
```

```python
a.isna().sum()
```

```python
 y=a['Survived']
 x=a.drop(['Survived'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.25)
```
